### 数据预处理，请移步baseline

```python
import pandas as pd
import pickle

with open('./data/trian.pkl', 'rb') as f:
    train = pickle.load(f)
with open('./data/test.pkl', 'rb') as f:
    test = pickle.load(f)
```

```python
## 对api进行词频统计
```

```python
def get_apis(df):
    group_filed = df.groupby('file_id')
    #按照tid 和 index 对api使用情况进行排序
    file_api = {}
    for file_id, file_group in group_filed:
        result = file_group.sort_values(['tid','index'], ascending=True)
        api_sequences = ' '.join(result['api'])
        file_api[file_id] = api_sequences
    return file_api
```

```python
train_apis = get_apis(train)
len(train_apis)
```

```python
test_apis = get_apis(test)
len(test_apis)
```

```python
## 拼接测试数据

with open('./data/df_train.pkl', 'rb') as f:
    df_train = pickle.load(f)
with open('./data/df_test.pkl', 'rb') as f:
    df_test = pickle.load(f)

df_train.drop(['api', 'tid', 'index'], axis=1, inplace=True)
df_test.drop(['api', 'tid', 'index'], axis=1, inplace=True)
```

```python
#拼接验证集
temp = pd.DataFrame.from_dict(train_apis, orient='index', columns=['api'])
temp = temp.reset_index().rename(columns={'index': 'file_id'})
df_train = df_train.merge(temp, on='file_id', how='left')

#拼接测试集
temp = pd.DataFrame.from_dict(test_apis, orient='index', columns=['api'])
temp = temp.reset_index().rename(columns={'index': 'file_id'})
df_test = df_test.merge(temp, on='file_id', how='left')
```

```python
## 在文本文件中，有一种可能就是在测试集中地单词未在训练集中出现，所有要和一下
df_all = pd.concat([df_train,df_test], axis=0)
```

```python
df_all['file_id'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(ngram_range=(1,3), min_df=0.01)
api_features =  vec.fit_transform(df_all['api'])
```

```python
# 将itidf创造成新的列

df_apis = pd.DataFrame(api_features.toarray(), columns=vec.get_feature_names_out())
```

```python
df_apis.to_pickle('./data/df_apis_tfidf.pkl')
```

```python
df_apis.shape
```

```python
## 分离开训练集和测试集地apis
df_trian_apis = df_apis[df_apis.index <= 13886]
df_test_apis = df_apis[df_apis.index > 13886]
```

```python
df_test_apis.index = range(len(df_test_apis))
```

```python
# 将数据合并
df_train = df_train.merge(df_trian_apis, left_index=True, right_index=True)
df_test = df_test.merge(df_test_apis, left_index=True, right_index=True)
```

```python
# 权重融合
def weight_fusion(df1, df2, weight1, weight2):
    result_weight = pd.DataFrame()
    for col in df1.columns:
        if col == 'file_id':
            result_weight['file_id'] = df1['file_id']
        else:
            result_weight[col] = df1[col] * weight1 + df2[col] * weight2
    return result_weight
```

```python
result  = weight_fusion(result_lgb_ifidf,result_xgb_ifidf, 0.5, 0.5)
```
